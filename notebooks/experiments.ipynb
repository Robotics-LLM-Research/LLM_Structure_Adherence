```python
import os
from dotenv import load_dotenv

from scripts.run_step_sequence import EXPERIMENTS, experiment
from src.utils import format_run_timestamp

RUN_ID = format_run_timestamp()
print(f"Results run_id: {RUN_ID}", flush=True)

load_dotenv()
TOKEN = os.getenv("HF_TOKEN")


def run_step_experiment(exp_idx: int) -> None:
    base_config = EXPERIMENTS[exp_idx]
    exp_config = {
        **base_config,
        "prefix": "Steps",
        "run_id": RUN_ID,
        "token": TOKEN,
    }
    print(
        f"Running experiment {exp_idx + 1}/{len(EXPERIMENTS)}: "
        f"{exp_config['model_id']} | uses_image={exp_config['uses_image']}",
        flush=True,
    )
    experiment(exp_config)


for idx, config in enumerate(EXPERIMENTS, start=1):
    print(
        f"{idx}. {config['model_id']} | uses_tools={config['uses_tools']} | "
        f"uses_image={config['uses_image']} | prompt_ids={config['prompt_ids']}",
        flush=True,
    )
```

```python
run_step_experiment(0)
```
