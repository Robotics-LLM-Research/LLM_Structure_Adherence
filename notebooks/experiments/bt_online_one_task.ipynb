```python
import sys
import traceback
from pathlib import Path

ROOT_DIR = Path.cwd().resolve().parents[1]
if str(ROOT_DIR) not in sys.path:
    sys.path.append(str(ROOT_DIR))

from scripts.run_bt_tasks_online import main as run_bt_tasks_online

EXP_ID = "bt_online_one_task_local-1"

def main(models):
    for model_id in models:
        print(f"===== ONE-TASK START: {model_id} =====", flush=True)
        try:
            run_bt_tasks_online(
                model_id=model_id,
                tasks_idx=[0],
                exp_id=EXP_ID,
            )
        except Exception as exc:
            print(f"ONE-TASK FAILED: {model_id}: {exc}", flush=True)
            traceback.print_exc()
            continue
        print(f"ONE-TASK DONE: {model_id}", flush=True)
```
